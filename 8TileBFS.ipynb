```python
class State:
    def __init__(self, board, parentMove):
        self.board = board
        self.parentMove = parentMove

    def display(self):
        i = 0
        while(i < 3):
          print(self.board[i])
          i+=1
        print('')

    def checkGoalReached(self):
        goalState = [[1,2,3],[8,0,4],[7,6,5]]
        for row in range(len(goalState)):
            for col in range(len(goalState[row])):
                if(goalState[row][col] != self.board[row][col]):
                    return False
        return True

    def deepCopyBoard(self):
        copy = [[],[],[]]
        for row in range(len(self.board)):
            for col in range(len(self.board[row])):
                copy[row].append(self.board[row][col])
        return copy

    def returnChildren(self):
        children = []

        zRow = 0
        zCol = 0
        found = False
        while((zRow < 3) and (found == False)):
          zCol = 0
          while((zCol < 3) and (found == False)):
            if(self.board[zRow][zCol] == 0):
              found = True
              break
            zCol+=1
          if(found == False):
            zRow+=1

        #IF NOT TOP ROW, THEN CAN MOVE DOWN
        if(zRow < 2):
          #we do not generate states that move back to grandParent
          if(self.parentMove != 'u'):
            down = self.deepCopyBoard()

            temp = down[zRow + 1][zCol]
            down[zRow + 1][zCol] = down[zRow][zCol]
            down[zRow][zCol] = temp

            children.append(State(down, 'd'))

        #Can move up
        if(zRow > 0):
          #we do not go back to parent
          if(self.parentMove != 'd'):
            #copy
            up = self.deepCopyBoard()
            #swap
            temp = up[zRow - 1][zCol]
            up[zRow - 1][zCol] = up[zRow][zCol]
            up[zRow][zCol] = temp
            #add to ren list
            children.append(State(up, 'u'))

        #Can move left
        if(zCol > 0):
          if(self.parentMove != 'r'):
            left = self.deepCopyBoard()
            temp = left[zRow][zCol - 1]
            left[zRow][zCol - 1] = left[zRow][zCol]
            left[zRow][zCol] = temp
            children.append(State(left, 'l'))

        #Can move right
        if(zCol < 2):
          if(self.parentMove != 'l'):
            right = self.deepCopyBoard()
            temp = right[zRow][zCol + 1]
            right[zRow][zCol + 1] = right[zRow][zCol]
            right[zRow][zCol] = temp
            children.append(State(right, 'r'))

        return children

visited = [] # List for visited nodes.
queue = [] #Initialize a queue

def checkVisited(child, visited):
    for state in visited:
        found = True
        for row in range(len(state)):
            for col in range(len(state[row])):
                if(child[row][col] != state[row][col]):
                    found = False
        if(found == True):
            print("Found a visited")
            return True
    return False


def bfs(visited, state):
    visited.append(state.board)
    queue.append(state)

    #while Queue is not empty
    while queue:
        currentState = queue.pop(0)
        currentState.display()

        #end BFS if goal is reached
        if(currentState.checkGoalReached() == True):
            #clear visited list
            visited = []
            return

        #generating child nodes
        children = currentState.returnChildren()

        for child in children:
            #we only visit nodes not visited previously
            if (checkVisited(child.board, visited) == False):
                visited.append(child.board)
                queue.append(child)



def dfs(visited, state):
    if(checkVisited(state.board, visited) == False):
       state.display()
       visited.append(state.board)

       #end DFS if goal is reached (Base case)
       if(state.checkGoalReached() == True):
          #clear visited list
          visited = []
          return True

       children = state.returnChildren()

       for child in children:
          if(dfs(visited, child) == True):
             return True



startState = State([[2,8,1],[0,4,3],[7,6,5]], '')


#print("Following is the Breadth-First Search")
#bfs(visited, startState)

print("Following is the Depth-First Search")
dfs(visited, startState)
```
