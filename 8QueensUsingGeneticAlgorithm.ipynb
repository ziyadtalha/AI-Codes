```python
import random
from random import randint
```

```python
#generate solutions
solutions = []

for counter in range(100):
    solution = []
    for i in range(8):
        gene = randint(0, 7)
        solution.append(gene)
    solutions.append(solution)

solutions
```

```python
#since we have a square board
def checkDiagonal(row1, col1, row2, col2):
    if(row1 - col1 == row2 - col2):
        return True
    if (row1 + col1 == row2 + col2):
        return True
    return False

def fitnessFunction(solution):
    conflicts = 0
    for i in range(len(solution)):
        for j in range(0, len(solution)):
            #no use checking the same column
            if(i != j):
                if solution[i] == solution[j] or (checkDiagonal(solution[i], i, solution[j], j) == True):
                    conflicts += 1
    return conflicts
```

```python
iterations = [100, 500, 1000]

for iteration in iterations:

    rankedSolutions = []
    for solution in solutions:
        rankedSolutions.append((fitnessFunction(solution), solution))
        
    #lower fitness is better in this case since it means less clashes
    rankedSolutions.sort()
    
    print(f"=== Generation 0 best solution: ====")
    print(rankedSolutions[0])

    if(rankedSolutions[0][0] != 0):
        
        print('')
        print(iteration, ' iterations:')
        print('')
        
        for genNumber in range(1, iteration):
        
            #let's keep the top 10 best solutions
            newGen = (rankedSolutions[:10])
        
            i = 10
            while (i < len(solutions)):
                solA = rankedSolutions[i][1]
                solB = rankedSolutions[i + 1][1]
                
                #performing crossover
        
                crossoverChance = random.uniform(0, 1)
                
                if(crossoverChance < 0.6):
                    randGeneIndexes = random.sample(range(0, 8), 3)
                    for index in randGeneIndexes:
                        #swapping genes
                        temp = solA[index]
                        solA[index] = solB[index]
                        solB[index] = temp
                
        
                #performing mutations
                
                mutationChanceA = random.uniform(0, 1)
                
                if(mutationChanceA < 0.1):
                    randGeneIndexes = random.sample(range(0, 8), 3)
                    for index in randGeneIndexes:         
                        solA[index] = randint(0, 7)
    
                mutationChanceB = random.uniform(0, 1)
                
                if(mutationChanceB < 0.1):
                    randGeneIndexes = random.sample(range(0, 8), 3)
                    for index in randGeneIndexes:         
                        solB[index] = randint(0, 7)
        
                newGen.append((fitnessFunction(solA), solA))
                newGen.append((fitnessFunction(solB), solB))
        
                i+=2
        
            newGen.sort()
        
            rankedSolutions = newGen
        
            print(f"=== Generation {genNumber} best solution: ====")
            print(rankedSolutions[0])
        
            if(rankedSolutions[0][0] == 0):
                break
    else:
        break
```
